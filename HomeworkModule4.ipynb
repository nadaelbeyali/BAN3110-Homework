```python
# Problem 1: Sales Reporting ---
# Setup dummy data in a list of dictionaries
sales_data = [
    {"date": "2026-04-18", "product": "Laptop", "units": 2, "price": 1200, "region": "East"},
    {"date": "2026-04-18", "product": "Mouse", "units": 10, "price": 25, "region": "West"},
    {"date": "2026-04-19", "product": "Monitor", "units": 5, "price": 300, "region": "East"},
    {"date": "2026-04-19", "product": "Keyboard", "units": 0, "price": 50, "region": "South"}, # Test case for 0 units
    {"date": "2026-04-19", "product": "Desk Lamp", "units": 3, "price": 45, "region": "West"}
]

total_revenue = 0
total_units = 0
high_value_threshold = 500

print("--- DAILY SALES REPORT SUMMARY ---")

# For loop to iterate through data
for sale in sales_data:
    # Use 'continue' to skip invalid entries (like 0 units sold)
    if sale["units"] <= 0:
        print(f"Skipping record: {sale['product']} (No units sold)")
        continue

    # Calculation logic
    transaction_total = sale["units"] * sale["price"]
    total_revenue += transaction_total
    total_units += sale["units"]

    # If statement for business logic (High-value flagging)
    if transaction_total >= high_value_threshold:
        status = "[HIGH VALUE]"
    else:
        status = "[Standard]"

    print(f"{sale['date']} | {sale['product']} | {sale['region']} | Total: ${transaction_total} {status}")

# While loop simulation for data auditing
audit_count = 0
while audit_count < len(sales_data):
    if sales_data[audit_count]["price"] > 2000:
        print("ALERT: Unusual price detected during audit!")
        break # Exit loop if something crazy is found
    audit_count += 1

print("-" * 30)
print(f"Grand Total Revenue: ${total_revenue}")
print(f"Total Units Moved: {total_units}")
```

```python
# Problem 3: Overdue Payments ---
from datetime import date

invoices = [
    ["INV-101", "Acme Corp", "2026-04-01", 1500.00, "Unpaid"],
    ["INV-102", "Global Inc", "2026-04-20", 2200.00, "Unpaid"],
    ["INV-103", "Stark Ind", "2026-03-15", 5000.00, "Unpaid"],
    ["INV-104", "Wayne Ent", "2026-04-10", 300.00, "Paid"],
    ["INV-105", "Oscorp", "2026-04-05", 1200.00, "Unpaid"]
]

today = date(2026, 4, 19)
overdue_balance = 0
follow_up_list = []

print(f"Status Report as of: {today}")
print(f"{'Invoice':<10} | {'Customer':<12} | {'Days Late':<10} | {'Status'}")

for inv in invoices:
    # Logic: Skip Paid Invoices
    if inv[4] == "Paid":
        continue

    # Manual date parsing logic (Simplified for beginner Python)
    year, month, day = map(int, inv[2].split("-"))
    due_date = date(year, month, day)

    if due_date < today:
        days_overdue = (today - due_date).days
        overdue_balance += inv[3]
        follow_up_list.append(inv[1])

        # Nested if for severity
        priority = "URGENT" if days_overdue > 20 else "Routine"
        print(f"{inv[0]:<10} | {inv[1]:<12} | {days_overdue:<10} | {priority}")

# Use a while loop to print the follow-up queue
print("\n--- Collections Call Queue ---")
counter = 0
while counter < len(follow_up_list):
    print(f"Call #{counter + 1}: {follow_up_list[counter]}")
    counter += 1
    if counter == 3: # Limit calls to top 3 for the morning shift
        print("...Queue paused for lunch break.")
        break

print(f"\nTotal Risk Exposure: ${overdue_balance}")
```

```python
# Problem 4: Data Cleaning ---
raw_customers = [
    "  john DOE  ", "alice_smith@email.com", "  WEST  ", 500,
    "jane sARAH ", "jane@business.org", "east", 1200,
    " BOB bobby ", "bob@web.com", "North ", "ERROR", # Error value to handle
    "clark Kent", "ck@dailyplanet.com", "West", 2500
]

cleaned_data = []
index = 0

# Using while loop to process a flat list in chunks of 4
while index < len(raw_customers):
    name = str(raw_customers[index]).strip().title()
    email = str(raw_customers[index+1]).lower()
    region = str(raw_customers[index+2]).strip().upper()
    sales = raw_customers[index+3]

    # Handle non-numeric sales errors
    if not isinstance(sales, (int, float)):
        print(f"Warning: Invalid sales value for {name}. Setting to 0.")
        sales = 0

    # If statement to flag high-value regions
    if region == "WEST":
        region_status = "Primary Territory"
    else:
        region_status = "Secondary Territory"

    cleaned_entry = [name, email, region, sales]
    cleaned_data.append(cleaned_entry)

    index += 4 # Move to next record

print("\n--- CLEANED CUSTOMER DATABASE ---")
for person in cleaned_data:
    # Use continue if email is missing (simple validation)
    if "@" not in person[1]:
        continue
    print(f"Customer: {person[0]} | Email: {person[1]} | Region: {person[2]} | Sales: ${person[3]}")
```

```python
# Problem 6: Shift Scheduling ---
available_staff = [
    {"name": "Alice", "day": "Monday", "shift": "Morning"},
    {"name": "Bob", "day": "Monday", "shift": "Evening"},
    {"name": "Charlie", "day": "Tuesday", "shift": "Morning"},
    {"name": "Diana", "day": "Monday", "shift": "Morning"}
]

needed_shifts = [
    {"day": "Monday", "shift": "Morning", "slots": 2},
    {"day": "Monday", "shift": "Evening", "slots": 1},
    {"day": "Tuesday", "shift": "Morning", "slots": 2}
]

schedule = []

print("--- GENERATING DRAFT SCHEDULE ---")

for need in needed_shifts:
    filled_slots = 0
    assigned_names = []

    for staff in available_staff:
        # Match day and shift
        if staff["day"] == need["day"] and staff["shift"] == need["shift"]:
            assigned_names.append(staff["name"])
            filled_slots += 1

            # Stop if we fill the slots needed
            if filled_slots == need["slots"]:
                break

    # Check if we are understaffed
    if filled_slots < need["slots"]:
        gap = need["slots"] - filled_slots
        status = f"!!! UNDERSTAFFED (Need {gap} more)"
    else:
        status = "Fully Staffed"

    print(f"{need['day']} {need['shift']}: {assigned_names} -> {status}")

# Final summary using a while loop
print("\nReviewing schedule completion...")
i = 0
while i < len(needed_shifts):
    print(f"Shift {i+1} verified.")
    i += 1
```
