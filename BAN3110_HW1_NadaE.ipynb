```python
#Problem1
import math
import random

u = random.random()
v = random.random() # random between 0 and 1

# Box-Muller formula: Z = sin(2*pi*v) * sqrt(-2 * ln(u))
z = math.sin(2 * math.pi * v) * math.sqrt(-2 * math.log(u))

print(f"Generated Gaussian value Z: {z}")
```

```python
#Problem3
import sys
sys.argv = ['test.py', '1.5', '2.5', '3.5']

x = float(sys.argv[1])
y = float(sys.argv[2])
z = float(sys.argv[3])

result = (x < y < z) or (x > y > z)
print(result)
```

```python
#Problem4
mag = float(input("Enter earthquake magnitude: "))

if mag < 2.0: desc = "Micro"
elif mag < 3.0: desc = "Very minor"
elif mag < 4.0: desc = "Minor"
elif mag < 5.0: desc = "Light"
elif mag < 6.0: desc = "Moderate"
elif mag < 7.0: desc = "Strong"
elif mag < 8.0: desc = "Major"
elif mag < 10.0: desc = "Great"
else: desc = "Meteoric"

print(f"A magnitude {mag} earthquake is considered to be a {desc} earthquake.")
```
