```python
#Problem1
count = 0
total_sum = 0
product = 1
sum_squares = 0

current_num = 150

while current_num <= 1000:
    if current_num % 2 != 0:
        count += 1
        total_sum += current_num
        product *= current_num
        sum_squares += current_num ** 2
    current_num += 1

print(f"a) Count: {count}")
print(f"b) Sum: {total_sum}")
print(f"c) Multiplication: {product}")
print(f"d) Sum of Squares: {sum_squares}")
```

```python
#Problem2
count = 0
total_sum = 0
product = 1
sum_squares = 0

for num in range(150, 1001): # 1001 to include 1000
    if num % 2 != 0:
        count += 1
        total_sum += num
        product *= num
        sum_squares += num ** 2

print(f"a) Count: {count}")
print(f"b) Sum: {total_sum}")
print(f"c) Multiplication: {product}")
print(f"d) Sum of Squares: {sum_squares}")
```

```python
#Problem3
# 1. Get user input and convert it to an integer
M = int(input("Enter a value for M: "))

# 2. Initialize the sum (s) and the counter (k)
s = 0
k = 1

# 3. Use the while-loop to compute the series
while k <= M:
    s = s + (1/k)  # We add the reciprocal of k to match the formula
    k = k + 1      # Increment the counter to move to the next fraction

# 4. Output the final result
print(f"The sum of the series for M = {M} is: {s}")
```

```python
#Problem4
total = 0
count = 0

while True:
    user_input = input("Enter a number: ")
    if user_input.lower() == "done":
        break
    try:
        num = float(user_input)
        total += num
        count += 1
    except ValueError:
        print("Invalid input")

if count > 0:
    print(total, count, total / count)
else:
    print("No numbers were entered.")
```

```python
#Problem5
maximum = None
minimum = None

while True:
    user_input = input("Enter a number: ")
    if user_input.lower() == "done":
        break
    try:
        num = float(user_input)
        if maximum is None or num > maximum:
            maximum = num
        if minimum is None or num < minimum:
            minimum = num
    except ValueError:
        print("Invalid input")

print(f"Maximum: {maximum}, Minimum: {minimum}")
```
